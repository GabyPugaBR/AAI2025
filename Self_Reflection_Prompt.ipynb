3.Self-Reflection Prompt for Improving Output

Goal: Ask the AI to critique and improve its summary.



Prompt 1: You are an agentic AI specialist. Built me a model that can be tested on Google Colab and follows the Self-Reflection Prompt for Improving Output framework.
It should be "Reflection prompt includes explicit criteria (accuracy, clarity, audience, length limits, formatting) and requires a revised version."
Here's the definition I have for that kind of prompt:
"Self-Reflection Prompts: Self-reflection prompts encourage the AI to evaluate and
critique its own output for accuracy, alignment with user goals, or clarity. This leads to
better refinement in a single interaction.
Example: Asking the agent to evaluate its outcome against the user's goal.
"Here's your answer to the user's question. Does your response meet the
requirement to be under 150 words and focused on data privacy concerns? If not,
revise it.""

```python
# Self-Reflection AI Agent
# Framework: Self-Reflection Prompt for Improving Output

#!pip install google-generativeai

from google.colab import userdata
import google.generativeai as genai

# --- Setup --- I added and edited API Key and model
api_key = userdata.get('Gemini_API_Key')
genai.configure(api_key=api_key)
model = genai.GenerativeModel("gemini-2.5-flash")

# -------------------------------------------------------
# STEP 1: Define the Reflection Criteria
# These are the explicit standards the AI must evaluate against
# -------------------------------------------------------
REFLECTION_CRITERIA = """
Evaluate the response against ALL of the following criteria:

1. ‚úÖ ACCURACY     ‚Äî Is the information factually correct and free of assumptions?
2. ‚úÖ CLARITY      ‚Äî Is the response easy to understand with no jargon or ambiguity?
3. ‚úÖ AUDIENCE     ‚Äî Is the tone and complexity appropriate for a general (non-technical) audience?
4. ‚úÖ LENGTH       ‚Äî Is the response under 150 words? Count carefully.
5. ‚úÖ FORMATTING   ‚Äî Does it use plain sentences? No bullet points, no headers, no markdown.

For each criterion, state: PASS or FAIL and briefly explain why.
Then provide a REVISED VERSION that fixes all failures.
Format your reflection exactly like this:

--- REFLECTION ---
ACCURACY: [PASS/FAIL] ‚Äî [reason]
CLARITY: [PASS/FAIL] ‚Äî [reason]
AUDIENCE: [PASS/FAIL] ‚Äî [reason]
LENGTH: [PASS/FAIL] ‚Äî [reason] (word count: X)
FORMATTING: [PASS/FAIL] ‚Äî [reason]

--- REVISED RESPONSE ---
[Your improved response here]
"""

# -------------------------------------------------------
# STEP 2: Initial Response Generation
# The AI answers the user's question without constraints first
# -------------------------------------------------------
def generate_initial_response(user_question: str) -> str:
    prompt = f"""You are a helpful AI assistant. Answer the following question:

Question: {user_question}

Give a thorough answer."""

    response = model.generate_content(prompt)
    return response.text


# -------------------------------------------------------
# STEP 3: Self-Reflection Pass
# The AI evaluates its own output against the criteria
# and produces a revised version
# -------------------------------------------------------
def reflect_and_revise(user_question: str, initial_response: str) -> str:
    reflection_prompt = f"""You are an AI output evaluator.

The user asked: "{user_question}"

Here is the AI's initial response:
\"\"\"
{initial_response}
\"\"\"

{REFLECTION_CRITERIA}
"""
    reflection = model.generate_content(reflection_prompt)
    return reflection.text


# -------------------------------------------------------
# STEP 4: Extract the Final Revised Response
# Parses out just the revised section for clean display
# -------------------------------------------------------
def extract_revised_response(reflection_output: str) -> str:
    marker = "--- REVISED RESPONSE ---"
    if marker in reflection_output:
        return reflection_output.split(marker)[1].strip()
    return "Could not extract revised response. See full reflection above."


# -------------------------------------------------------
# STEP 5: Main Runner ‚Äî Full Self-Reflection Pipeline
# -------------------------------------------------------
def run_self_reflection_agent(user_question: str):
    print("=" * 60)
    print("ü§ñ SELF-REFLECTION AI AGENT")
    print("=" * 60)
    print(f"\nüìù USER QUESTION:\n{user_question}\n")

    # Stage 1: Generate raw response
    print("‚è≥ Stage 1: Generating initial response...\n")
    initial = generate_initial_response(user_question)
    print("--- INITIAL RESPONSE ---")
    print(initial)

    # Stage 2: Reflect and revise
    print("\n‚è≥ Stage 2: Running self-reflection against criteria...\n")
    reflection = reflect_and_revise(user_question, initial)
    print(reflection)

    # Stage 3: Surface the final clean answer
    print("\n" + "=" * 60)
    print("‚úÖ FINAL DELIVERABLE (Post-Reflection)")
    print("=" * 60)
    revised = extract_revised_response(reflection)
    print(revised)
    print("=" * 60)


# -------------------------------------------------------
# TEST CASES ‚Äî Try different topics to see the reflection
# -------------------------------------------------------

# Test 1: A topic likely to produce a too-long, too-technical response
run_self_reflection_agent(
    "What is artificial intelligence and how does it work?"
)
```

```python
# Testing with a different question
# I also changed this to use OpenAI's API and added enough credits to test it.

# Self-Reflection AI Agent
# Framework: Self-Reflection Prompt for Improving Output
# Tests on Google Colab using OpenAI API

!pip install openai

from google.colab import userdata
from openai import OpenAI

# --- Setup --- I added and edited API Key and model
api_key = userdata.get('OpenAI_API')
client = OpenAI(api_key=api_key)


# -------------------------------------------------------
# STEP 1: Define the Reflection Criteria
# These are the explicit standards the AI must evaluate against
# -------------------------------------------------------
REFLECTION_CRITERIA = """
Evaluate the response against ALL of the following criteria:

1. ‚úÖ ACCURACY     ‚Äî Is the information factually correct and free of assumptions?
2. ‚úÖ CLARITY      ‚Äî Is the response easy to understand with no jargon or ambiguity?
3. ‚úÖ AUDIENCE     ‚Äî Is the tone and complexity appropriate for a general (non-technical) audience?
4. ‚úÖ LENGTH       ‚Äî Is the response under 150 words? Count carefully.
5. ‚úÖ FORMATTING   ‚Äî Does it use plain sentences? No bullet points, no headers, no markdown.

For each criterion, state: PASS or FAIL and briefly explain why.
Then provide a REVISED VERSION that fixes all failures.
Format your reflection exactly like this:

--- REFLECTION ---
ACCURACY: [PASS/FAIL] ‚Äî [reason]
CLARITY: [PASS/FAIL] ‚Äî [reason]
AUDIENCE: [PASS/FAIL] ‚Äî [reason]
LENGTH: [PASS/FAIL] ‚Äî [reason] (word count: X)
FORMATTING: [PASS/FAIL] ‚Äî [reason]

--- REVISED RESPONSE ---
[Your improved response here]
"""

# -------------------------------------------------------
# STEP 2: Initial Response Generation
# The AI answers the user's question without constraints first
# -------------------------------------------------------
def generate_initial_response(user_question: str) -> str:
    response = client.chat.completions.create(
        model="gpt-4.1-nano",
        messages=[
            {"role": "system", "content": "You are a helpful AI assistant."},
            {"role": "user", "content": f"Answer the following question thoroughly:\n\n{user_question}"}
        ]
    )
    return response.choices[0].message.content



# -------------------------------------------------------
# STEP 3: Self-Reflection Pass
# The AI evaluates its own output against the criteria
# and produces a revised version
# -------------------------------------------------------
def reflect_and_revise(user_question: str, initial_response: str) -> str:
    reflection_prompt = f"""You are an AI output evaluator.

The user asked: "{user_question}"

Here is the AI's initial response:
\"\"\"
{initial_response}
\"\"\"

{REFLECTION_CRITERIA}
"""
    response = client.chat.completions.create(
        model="gpt-4.1-nano",
        messages=[
            {"role": "user", "content": reflection_prompt}
        ]
    )
    return response.choices[0].message.content


# -------------------------------------------------------
# STEP 4: Extract the Final Revised Response
# Parses out just the revised section for clean display
# -------------------------------------------------------
def extract_revised_response(reflection_output: str) -> str:
    marker = "--- REVISED RESPONSE ---"
    if marker in reflection_output:
        return reflection_output.split(marker)[1].strip()
    return "Could not extract revised response. See full reflection above."


# -------------------------------------------------------
# STEP 5: Main Runner ‚Äî Full Self-Reflection Pipeline
# -------------------------------------------------------
def run_self_reflection_agent(user_question: str):
    print("=" * 60)
    print("ü§ñ SELF-REFLECTION AI AGENT")
    print("=" * 60)
    print(f"\nüìù USER QUESTION:\n{user_question}\n")

    # Stage 1: Generate raw response
    print("‚è≥ Stage 1: Generating initial response...\n")
    initial = generate_initial_response(user_question)
    print("--- INITIAL RESPONSE ---")
    print(initial)

    # Stage 2: Reflect and revise
    print("\n‚è≥ Stage 2: Running self-reflection against criteria...\n")
    reflection = reflect_and_revise(user_question, initial)
    print(reflection)

    # Stage 3: Surface the final clean answer
    print("\n" + "=" * 60)
    print("‚úÖ FINAL DELIVERABLE (Post-Reflection)")
    print("=" * 60)
    revised = extract_revised_response(reflection)
    print(revised)
    print("=" * 60)


# -------------------------------------------------------
# TEST CASES ‚Äî Try different topics to see the reflection
# -------------------------------------------------------

# Test 1: A topic likely to produce a too-long, too-technical response
run_self_reflection_agent(
    "Summarize in 2 sentences the trading Company Robinhood"
)
```
